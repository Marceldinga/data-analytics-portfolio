```python
# ✅ COMPLETE SINGLE CODE — Load partitioned CSVs + Train + Validate Logistic Regression

import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix

# -----------------------------
# 1) Load partitioned datasets
# -----------------------------
train_df = pd.read_csv("train_dataset.csv")
val_df   = pd.read_csv("validation_dataset.csv")
test_df  = pd.read_csv("test_dataset.csv")  # loaded (not used below)

print("Train:", train_df.shape, "Val:", val_df.shape, "Test:", test_df.shape)

# -----------------------------
# 2) Split into X/y
# -----------------------------
TARGET = "default"

y_train = train_df[TARGET].astype(int)
X_train = train_df.drop(columns=[TARGET])

y_val = val_df[TARGET].astype(int)
X_val = val_df.drop(columns=[TARGET])

print("Default rates:", round(y_train.mean(),4), round(y_val.mean(),4))

# -----------------------------
# 3) Scale features (fit on train only)
# -----------------------------
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled   = scaler.transform(X_val)

# -----------------------------
# 4) Train Logistic Regression
# -----------------------------
log_model = LogisticRegression(
    max_iter=3000,
    class_weight="balanced"
)
log_model.fit(X_train_scaled, y_train)

# -----------------------------
# 5) Validate
# -----------------------------
y_val_pred = log_model.predict(X_val_scaled)
y_val_prob = log_model.predict_proba(X_val_scaled)[:, 1]

print("\n=== VALIDATION RESULTS ===")
print(classification_report(y_val, y_val_pred))
print("Validation ROC-AUC:", round(roc_auc_score(y_val, y_val_prob), 4))
print("Confusion Matrix:\n", confusion_matrix(y_val, y_val_pred))
```

```python
# ✅ COMPLETE SINGLE CODE — RandomForest optimized to CATCH MORE DEFAULTS (HIGH RECALL)
# Goal: maximize Recall for class 1 by choosing a threshold that achieves TARGET_RECALL

import numpy as np
import matplotlib.pyplot as plt

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    classification_report, confusion_matrix,
    roc_auc_score, roc_curve,
    precision_recall_curve, f1_score, recall_score, precision_score
)

# -----------------------------
# 0) Assumes you already have:
#    X_train, y_train, X_val, y_val
# -----------------------------
print("Train:", getattr(X_train, "shape", None), "Val:", getattr(X_val, "shape", None))
print(f"Default rates: {np.mean(y_train):.4f} {np.mean(y_val):.4f}")

# -----------------------------
# 1) OPTIONAL speed sample (set SAMPLE_N=None to disable)
# -----------------------------
SAMPLE_N = 250_000  # increase if you want (e.g., 400000) or set None for full
if SAMPLE_N is not None and X_train.shape[0] > SAMPLE_N:
    rng = np.random.default_rng(42)
    idx = rng.choice(X_train.shape[0], size=SAMPLE_N, replace=False)
    X_tr = X_train.iloc[idx] if hasattr(X_train, "iloc") else X_train[idx]
    y_tr = y_train.iloc[idx] if hasattr(y_train, "iloc") else y_train[idx]
    print("Using sample for training:", X_tr.shape)
else:
    X_tr, y_tr = X_train, y_train
    print("Using full training data.")

# -----------------------------
# 2) Train RandomForest (stable/fast settings)
# -----------------------------
rf = RandomForestClassifier(
    n_estimators=500,
    max_depth=16,
    min_samples_leaf=30,
    min_samples_split=80,
    max_features="sqrt",
    n_jobs=-1,
    random_state=42,
    class_weight="balanced_subsample",
    verbose=1
)

rf.fit(X_tr, y_tr)

# -----------------------------
# 3) Predict probabilities on validation
# -----------------------------
val_proba = rf.predict_proba(X_val)[:, 1]
auc = roc_auc_score(y_val, val_proba)
print("\nValidation ROC-AUC:", round(auc, 4))

# -----------------------------
# 4) Helper to evaluate at threshold
# -----------------------------
def eval_threshold(y_true, proba, thr, label=""):
    pred = (proba >= thr).astype(int)
    cm = confusion_matrix(y_true, pred)
    prec = precision_score(y_true, pred, zero_division=0)
    rec  = recall_score(y_true, pred, zero_division=0)
    f1   = f1_score(y_true, pred, zero_division=0)
    if label:
        print(f"\n=== {label} ===")
    print(f"Threshold: {thr:.2f} | Precision(1): {prec:.4f} | Recall(1): {rec:.4f} | F1(1): {f1:.4f}")
    print("Confusion Matrix:\n", cm)
    print(classification_report(y_true, pred, digits=4, zero_division=0))
    return pred, cm, prec, rec, f1

# Baseline at 0.50
_ = eval_threshold(y_val, val_proba, 0.50, "BASELINE (0.50)")

# -----------------------------
# 5) Choose threshold to HIT TARGET RECALL (catch more defaults)
#    We pick the HIGHEST threshold that still meets recall >= TARGET_RECALL
#    (highest threshold = fewer false positives while meeting recall goal)
# -----------------------------
TARGET_RECALL = 0.70  # <- change to 0.75 or 0.80 if you want to catch even more

grid = np.linspace(0.01, 0.99, 199)

recs = []
precs = []
f1s = []

for t in grid:
    p = (val_proba >= t).astype(int)
    recs.append(recall_score(y_val, p, zero_division=0))
    precs.append(precision_score(y_val, p, zero_division=0))
    f1s.append(f1_score(y_val, p, zero_division=0))

eligible = [t for t, r in zip(grid, recs) if r >= TARGET_RECALL]

print("\n=== THRESHOLD SEARCH (HIGH RECALL) ===")
if not eligible:
    print(f"Could not reach Recall >= {TARGET_RECALL:.2f}.")
    # fallback: pick threshold with max recall
    best_i = int(np.argmax(recs))
    best_thr = float(grid[best_i])
    print("Fallback threshold (max recall):", round(best_thr, 2), "Recall:", round(recs[best_i], 4))
else:
    best_thr = float(max(eligible))
    i = list(grid).index(best_thr)
    print(f"Target Recall: {TARGET_RECALL:.2f} | Chosen Threshold: {best_thr:.2f}")
    print(f"At chosen threshold -> Recall: {recs[i]:.4f}, Precision: {precs[i]:.4f}, F1: {f1s[i]:.4f}")

final_pred, final_cm, final_prec, final_rec, final_f1 = eval_threshold(
    y_val, val_proba, best_thr, f"FINAL (TARGET RECALL={TARGET_RECALL:.2f})"
)

# -----------------------------
# 6) Plot ROC curve
# -----------------------------
fpr, tpr, _ = roc_curve(y_val, val_proba)
plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, label=f"AUC = {auc:.4f}")
plt.plot([0,1],[0,1],'--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve (Validation)")
plt.legend()
plt.show()

# -----------------------------
# 7) Plot Precision-Recall curve
# -----------------------------
prec_curve, rec_curve, _ = precision_recall_curve(y_val, val_proba)
plt.figure(figsize=(6,5))
plt.plot(rec_curve, prec_curve)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve (Validation)")
plt.show()

# -----------------------------
# 8) Business impact estimate (simple)
# -----------------------------
AVG_LOAN = 15000   # edit to your dataset context
LOSS_RATE = 0.70   # portion lost when default happens (after recovery)

FN = final_cm[1, 0]   # missed defaults (bad)
TP = final_cm[1, 1]   # caught defaults (good)
FP = final_cm[0, 1]   # false alarms (reject good)
TN = final_cm[0, 0]

estimated_loss = FN * AVG_LOAN * LOSS_RATE

print("\n=== BUSINESS IMPACT (SIMPLE) ===")
print("Missed defaults (FN):", FN)
print("Caught defaults (TP):", TP)
print("False alarms (FP):", FP)
print(f"Estimated loss from missed defaults ≈ ${estimated_loss:,.0f} (AVG_LOAN={AVG_LOAN}, LOSS_RATE={LOSS_RATE})")

# -----------------------------
# 9) Optional: show top features
# -----------------------------
try:
    import pandas as pd
    if hasattr(X_train, "columns"):
        fi = pd.DataFrame({
            "feature": X_train.columns,
            "importance": rf.feature_importances_
        }).sort_values("importance", ascending=False)
        print("\n=== TOP 20 FEATURES (RF IMPORTANCE) ===")
        print(fi.head(20).to_string(index=False))
except Exception as e:
    print("\nFeature importance skipped:", e)
```

```python
# ✅ COMPLETE SINGLE CODE — XGBoost (Compatible Version)

import numpy as np
import matplotlib.pyplot as plt
import xgboost as xgb

from sklearn.metrics import (
    classification_report, confusion_matrix,
    roc_auc_score, roc_curve,
    precision_recall_curve,
    precision_score, recall_score, f1_score
)

# -------------------------------------------------
# Assume X_train, y_train, X_val, y_val exist
# -------------------------------------------------

print("Train:", X_train.shape, "Val:", X_val.shape)
print(f"Default rates: {np.mean(y_train):.4f} {np.mean(y_val):.4f}")

# -------------------------------------------------
# Handle class imbalance
# -------------------------------------------------
pos = np.sum(y_train)
neg = len(y_train) - pos
scale_pos_weight = neg / pos
print("scale_pos_weight:", round(scale_pos_weight, 4))

# -------------------------------------------------
# Train XGBoost (no early stopping)
# -------------------------------------------------
model = xgb.XGBClassifier(
    n_estimators=800,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.85,
    colsample_bytree=0.85,
    scale_pos_weight=scale_pos_weight,
    objective="binary:logistic",
    eval_metric="auc",
    tree_method="hist",
    n_jobs=-1,
    random_state=42
)

model.fit(X_train, y_train)

# -------------------------------------------------
# Predict probabilities
# -------------------------------------------------
val_proba = model.predict_proba(X_val)[:, 1]
auc = roc_auc_score(y_val, val_proba)
print("\nValidation ROC-AUC:", round(auc, 4))

# -------------------------------------------------
# Baseline 0.50 threshold
# -------------------------------------------------
def evaluate(threshold):
    preds = (val_proba >= threshold).astype(int)
    cm = confusion_matrix(y_val, preds)
    prec = precision_score(y_val, preds)
    rec  = recall_score(y_val, preds)
    f1   = f1_score(y_val, preds)

    print("\nThreshold:", round(threshold,4))
    print("Precision:", round(prec,4),
          "Recall:", round(rec,4),
          "F1:", round(f1,4))
    print("Confusion Matrix:\n", cm)
    print(classification_report(y_val, preds))
    return cm

print("\n=== BASELINE (0.50) ===")
evaluate(0.5)

# -------------------------------------------------
# Smart threshold for high recall
# -------------------------------------------------
TARGET_RECALL = 0.80

grid = np.linspace(0.01, 0.99, 300)

best = None

for t in grid:
    preds = (val_proba >= t).astype(int)
    rec = recall_score(y_val, preds)
    prec = precision_score(y_val, preds)
    f1 = f1_score(y_val, preds)

    if rec >= TARGET_RECALL:
        if best is None or prec > best[0]:
            best = (prec, t, rec, f1)

print("\n=== SMART THRESHOLD ===")

if best is None:
    print("Could not reach target recall.")
else:
    print("Chosen threshold:", round(best[1],4))
    print("Precision:", round(best[0],4),
          "Recall:", round(best[2],4),
          "F1:", round(best[3],4))
    final_cm = evaluate(best[1])

# -------------------------------------------------
# ROC Curve
# -------------------------------------------------
fpr, tpr, _ = roc_curve(y_val, val_proba)

plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, label=f"AUC = {auc:.4f}")
plt.plot([0,1],[0,1],'--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve (XGBoost)")
plt.legend()
plt.show()

# -------------------------------------------------
# Precision-Recall Curve
# -------------------------------------------------
prec_curve, rec_curve, _ = precision_recall_curve(y_val, val_proba)

plt.figure(figsize=(6,5))
plt.plot(rec_curve, prec_curve)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve (XGBoost)")
plt.show()

# -------------------------------------------------
# Business impact
# -------------------------------------------------
AVG_LOAN = 15000
LOSS_RATE = 0.70

FN = final_cm[1,0]
estimated_loss = FN * AVG_LOAN * LOSS_RATE

print("\n=== BUSINESS IMPACT ===")
print("Missed defaults (FN):", FN)
print("Estimated loss ≈ $", format(estimated_loss, ","))
```

```python
# ✅ COMPLETE SINGLE CODE — Train + Validate LOGISTIC REGRESSION (Catch More Defaults)
# Assumes you already have: X_train, y_train, X_val, y_val
# Works with pandas or numpy. Uses scaling. Tunes threshold for target recall.

import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    classification_report, confusion_matrix,
    roc_auc_score, roc_curve,
    precision_recall_curve,
    precision_score, recall_score, f1_score
)

# -----------------------------
# 0) Sanity checks
# -----------------------------
print("Train:", X_train.shape, "Val:", X_val.shape)
print(f"Default rates: {np.mean(y_train):.4f} {np.mean(y_val):.4f}")

# -----------------------------
# 1) Scale features (important for Logistic Regression)
# -----------------------------
scaler = StandardScaler(with_mean=True, with_std=True)
Xtr = scaler.fit_transform(X_train)
Xva = scaler.transform(X_val)

# -----------------------------
# 2) Train Logistic Regression (class_weight helps imbalance)
# -----------------------------
logit = LogisticRegression(
    solver="lbfgs",
    max_iter=2000,
    n_jobs=None,                 # lbfgs ignores n_jobs
    class_weight="balanced",     # helps catch more defaults
    random_state=42
)

logit.fit(Xtr, y_train)

# -----------------------------
# 3) Predict probabilities
# -----------------------------
val_proba = logit.predict_proba(Xva)[:, 1]
auc = roc_auc_score(y_val, val_proba)
print("\nValidation ROC-AUC:", round(auc, 4))

# -----------------------------
# 4) Evaluate helper
# -----------------------------
def evaluate_at_threshold(y_true, proba, thr, title=""):
    pred = (proba >= thr).astype(int)
    cm = confusion_matrix(y_true, pred)
    prec = precision_score(y_true, pred, zero_division=0)
    rec  = recall_score(y_true, pred, zero_division=0)
    f1   = f1_score(y_true, pred, zero_division=0)

    if title:
        print(f"\n=== {title} ===")
    print(f"Threshold: {thr:.4f} | Precision(1): {prec:.4f} | Recall(1): {rec:.4f} | F1(1): {f1:.4f}")
    print("Confusion Matrix:\n", cm)
    print(classification_report(y_true, pred, digits=4, zero_division=0))
    return pred, cm, prec, rec, f1

# Baseline threshold 0.50
_ = evaluate_at_threshold(y_val, val_proba, 0.50, title="LOGISTIC BASELINE (0.50)")

# -----------------------------
# 5) SMART threshold to catch more defaults:
#    choose threshold that achieves TARGET_RECALL while maximizing precision
# -----------------------------
TARGET_RECALL = 0.80  # change to 0.75 / 0.85 as needed
grid = np.linspace(0.01, 0.99, 400)

best = None  # (precision, threshold, recall, f1)
best_any = None  # fallback by max recall

for t in grid:
    pred = (val_proba >= t).astype(int)
    rec = recall_score(y_val, pred, zero_division=0)
    prec = precision_score(y_val, pred, zero_division=0)
    f1 = f1_score(y_val, pred, zero_division=0)

    if best_any is None or rec > best_any[2] or (rec == best_any[2] and prec > best_any[0]):
        best_any = (prec, float(t), rec, f1)

    if rec >= TARGET_RECALL:
        if best is None:
            best = (prec, float(t), rec, f1)
        else:
            if (prec > best[0]) or (prec == best[0] and f1 > best[3]) or (prec == best[0] and f1 == best[3] and t > best[1]):
                best = (prec, float(t), rec, f1)

print("\n=== SMART THRESHOLD (HIGH RECALL) ===")
if best is None:
    chosen_thr = best_any[1]
    print(f"Could not reach recall >= {TARGET_RECALL:.2f}. Using max-recall fallback.")
    print(f"Fallback threshold: {chosen_thr:.4f} | Precision: {best_any[0]:.4f} | Recall: {best_any[2]:.4f} | F1: {best_any[3]:.4f}")
else:
    chosen_thr = best[1]
    print(f"Target recall: {TARGET_RECALL:.2f}")
    print(f"Chosen threshold: {chosen_thr:.4f} | Precision: {best[0]:.4f} | Recall: {best[2]:.4f} | F1: {best[3]:.4f}")

final_pred, final_cm, final_prec, final_rec, final_f1 = evaluate_at_threshold(
    y_val, val_proba, chosen_thr, title=f"LOGISTIC FINAL (TARGET RECALL={TARGET_RECALL:.2f})"
)

# -----------------------------
# 6) ROC Curve
# -----------------------------
fpr, tpr, _ = roc_curve(y_val, val_proba)
plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, label=f"AUC = {auc:.4f}")
plt.plot([0, 1], [0, 1], "--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve (Logistic Regression)")
plt.legend()
plt.show()

# -----------------------------
# 7) Precision-Recall Curve
# -----------------------------
prec_curve, rec_curve, _ = precision_recall_curve(y_val, val_proba)
plt.figure(figsize=(6, 5))
plt.plot(rec_curve, prec_curve)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve (Logistic Regression)")
plt.show()

# -----------------------------
# 8) Business impact estimate (simple)
# -----------------------------
AVG_LOAN = 15000
LOSS_RATE = 0.70

FN = final_cm[1, 0]
estimated_loss = FN * AVG_LOAN * LOSS_RATE

print("\n=== BUSINESS IMPACT (LOGISTIC) ===")
print("Missed defaults (FN):", FN)
print(f"Estimated loss from missed defaults ≈ ${estimated_loss:,.0f} (AVG_LOAN={AVG_LOAN}, LOSS_RATE={LOSS_RATE})")

# -----------------------------
# 9) Optional: show top positive/negative coefficients (interpretability)
# -----------------------------
try:
    if hasattr(X_train, "columns"):
        import pandas as pd
        coefs = logit.coef_.ravel()
        coef_df = pd.DataFrame({"feature": X_train.columns, "coef": coefs})
        print("\n=== TOP 15 FEATURES INCREASING DEFAULT RISK (largest +coef) ===")
        print(coef_df.sort_values("coef", ascending=False).head(15).to_string(index=False))
        print("\n=== TOP 15 FEATURES DECREASING DEFAULT RISK (largest -coef) ===")
        print(coef_df.sort_values("coef", ascending=True).head(15).to_string(index=False))
except Exception as e:
    print("\nCoefficient display skipped:", e)
```
